```python
df1 = df.dropna(subset=['medal']) 
df1# Ensure we only use valid medal records
```

```python
df1['year'] = df1['year'].astype(int)
df1['year']
```

```python
#Total Medals per Olympic Year
yearly_medals = df['year'].value_counts().sort_index()
yearly_medals
```

```python
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.barplot(x=yearly_medals.index, y=yearly_medals.values, palette='viridis')
plt.title('Total Biathlon Medals Awarded per Olympic Year')
plt.xlabel('year')
plt.ylabel(' medal')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Country-Wise Medal Count
country_medals = df['country'].value_counts().reset_index()
country_medals.columns = ['country', 'medals']
print(country_medals)
print(country_medals.columns)
```

```python
#Country-wise Medal Performance Over Time
country_year = df.groupby(['year', 'country']).size().reset_index(name='medals')
pivot = country_year.pivot(index='year', columns='country', values='medals').fillna(0)
print(pivot)
```

```python
#Focus on top 5 countries only
top_countries = country_medals.head(5)['country']
pivot = pivot[top_countries]
pivot
```

```python
pivot.plot(marker='o', figsize=(14, 6))
plt.title('Top 5 Countries - Biathlon Medal Count Over Time')
plt.xlabel('Olympic Year')
plt.ylabel('Number of Medals')
plt.legend(title='Country')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Medals Per Capita (Mock Population Dictionary)
pop_dict = {
    'Norway': 5400000,
    'Germany': 83000000,
    'Russia': 144000000,
    'France': 67000000,
    'USA': 331000000,
   
}
```

```python
#Add per capita column
country_medals['population'] = country_medals['country'].map(pop_dict)
country_medals = country_medals.dropna(subset=['population'])  # Drop if population unknown
country_medals['medal'] = country_medals['medals'] / (country_medals['population'] / 1_000_000)
country_medals['medal']
```

```python
#Plot per capita
top_per_capita = country_medals.sort_values('medal', ascending=False).head(10)
plt.figure(figsize=(10, 5))
sns.barplot(data=top_per_capita, x='country', y='medal', palette='rocket')
plt.title('Top Countries by Biathlon Medals per Capita')
plt.ylabel('Medals per Million People')
plt.tight_layout()
plt.show()
```

```python
# Event-level breakdown (within Biathlon)
event_medals = df['event_title'].value_counts().reset_index()
event_medals.columns = ['Event', 'Medal_Count']
print(event_medals.head())

# Plot
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.barplot(data=event_medals, x='Medal_Count', y='Event', palette='coolwarm')
plt.title('Medal Count by Biathlon Event (1960â€“2022)')
plt.xlabel('Medals')
plt.ylabel('Event')
plt.tight_layout()
plt.show()
```
